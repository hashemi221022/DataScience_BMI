```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("insurance.csv")
data
```

```python
df = pd.DataFrame(data)
df
```

```python
plt.scatter(df["bmi"], df["age"])
plt.xlabel('bmi')
plt.ylabel('age')
plt.grid()
plt.savefig("bmi age")
plt.show()
```

```python
plt.scatter(df["sex"], df["bmi"])
plt.xlabel('sex')
plt.ylabel('bmi')
plt.grid()
plt.savefig("bmi sex")
plt.show()
```

```python
plt.scatter(df["children"], df["age"])
plt.xlabel('children')
plt.ylabel('age')
plt.grid()
plt.savefig("age children")
plt.show()
```
